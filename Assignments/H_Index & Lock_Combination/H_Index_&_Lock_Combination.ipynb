# **1. H_INDEX**
> # **a. Abstraction:**
>>  Tìm số K lớn nhất sao cho trong mảng A chứa N phần tử có K phần tử có giá trị lớn hơn K, biết K < N.

> # **b. Pattern Recognition:**
>> Kỹ thuật áp dụng: Duyệt mảng.

>> **Đặc điểm nhận dạng**:

>> Ví dụ cho mảng arr = [3, 4, 8, 6, 5, 2, 9, 7].

>>Với K=8 có 2 phần tử >= 8  => không thỏa.

>>Với K=7 có 3 phần tử >= 7  => không thỏa.

>>Với K=6 có 4 phần tử >= 6  => không thỏa.

>>Với K=5 có 5 phần tử >= 5  => thỏa.

>>Với K=4 có 6 phần tử >= 4  => thỏa.

>>Với K=3 có 7 phần tử >= 3  => thỏa.

>>Với K=2 có 8 phần tử >= 2  => thỏa.

>>Với K=1 có 8 phần tử >= 1  => thỏa.

# -> Vậy K = 6 thỏa yêu cầu của đề.

> # **c. Algorithm designed:**
```
def H_index(A):
    A.sort()
    for i, val in enumerate(A):
        res = len(A) - i
        if res <= val:
            return res
    return 0
n = int(input())
A = list(map(int,input().split()))
print(H_index(A))
```
> # **d. Evaluate Complexity of Solution**
>> Độ phức tạp: O(nlogn)

# **2. KHOÁ SỐ**
> # **a. Abstraction:**
>>  Cho một chuỗi số, tìm chuỗi số lớn nhất chia hết cho 3 bằng cách đổi chỗ vị trí hoặc xóa các ký tự trong chuỗi số đó.

> # **b. Pattern Recognition:**
>> Kỹ thuật áp dụng: Duyệt mảng, sắp xếp mảng.

>> **Đặc điểm nhận dạng**: Thay đổi hoặc xóa các vị trí trong chuỗi để được số nguyên lớn nhất chia hết cho 3.


> # **c. Algorithm designed:**
```
def Remove_Smallest(x,r,arr):
        idx = r.index(x)
        arr.pop(idx)
        r.pop(idx)
```
```
def Largest_Mul_Three(arr):
    r = [a % 3 for a in arr]
    s = sum(r) % 3
    if s != 0:
        try:
            Remove_Smallest(s, r, arr)
        except:
            Remove_Smallest(3-s, r, arr)
            Remove_Smallest(3-s, r, arr)
    if len(arr) == 0:
        print("")
    else:
        arr.sort(reverse = True)
        print(*arr,sep="")
```
**Main Block:**
```
from sys import stdin
string = list(stdin.readline().strip())
arr = []
for i in string:
    arr.append(int(i))
arr.sort()
Largest_Mul_Three(arr)
```
> # **d. Evaluate Complexity of Solution**
>> Độ phức tạp: O(n^2)
